```python
from random import shuffle

suits = ["Hearts", "Diamonds", "Spades", "Clubs"]
ranks = ["Two", "Three", "Fore", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace"]
values = {"Two": 2, "Three": 3, "Fore": 4, "Five": 5, "Six": 6, "Seven": 7, "Eight": 8, "Nine": 9, "Ten": 10,
          "Jack": 11, "Queen": 12, "King": 13, "Ace": 14}
```

```python
class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]

    def __str__(self):
        return f"{self.rank} of {self.suit}"
```

```python
class Deck:
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                temp = Card(suit, rank)
                self.all_cards.append(temp)

    def shuffle(self):
        shuffle(self.all_cards)

    def deal_one(self):
        return self.all_cards.pop()
```

```python
class Player:
    def __init__(self, name):
        self.name = name
        self.all_cards = []

    def remove_one(self):
        return self.all_cards.pop()

    def add_card(self, add_card):
        if type(add_card) == list:
            self.all_cards.extend(add_card)
        else:
            self.all_cards.append(add_card)

    def __str__(self):
        return f"{self.name} has a {len(self.all_cards)} cards"
```

```python
game_on=True
player_one = Player("One")
player_two = Player("Two")
new_deck = Deck()
new_deck.shuffle()
round=0
```

```python
for x in range(26):
    player_one.add_card(new_deck.deal_one())
    player_two.add_card(new_deck.deal_one())
```

```python
while game_on:
    round+=1
    print(f"Round {round}")
    if len(player_one.all_cards)==0:
        print("Player One has no Cards, Player Two Wins!!!")
        game_on=False
        break
    if len(player_two.all_cards)==0:
        print("Player Two has no Cards, Player One Wins!!!")
        game_on=False
        break
    player_one_holds=[]
    player_one_holds.append(player_one.remove_one())
    player_two_holds=[]
    player_two_holds.append(player_two.remove_one())
    at_war=True
    while at_war:
        if player_one_holds[-1].value>player_two_holds[-1].value:
            player_one.all_cards[player_one_holds]
            player_one.all_cards[player_two_holds]
            at_war=False
            break
        elif player_one_holds[-1].value<player_two_holds[-1].value:
            player_two.all_cards[player_two_holds]
            player_two.all_cards[player_one_holds]
            at_war=False
            break
        else:
            print("WAR!!!")
            if len(player_one.all_cards)<5:
                print("Player One lose, Player TWO Wins the WAR!!")
                game_on=False
                break
            if len(player_two.all_cards)<5:
                print("Player two lose, Player ONE Wins the WAR!!")
                game_on=False
                break


```
